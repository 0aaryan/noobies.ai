```python
def transcribe_audio(audio_path , word_timestamps = True):
    import whisper
    model = whisper.load_model("base.en")
    result = model.transcribe(
        audio_path,
        word_timestamps=word_timestamps,

    )
    words = []
    for res in result["segments"]:
        words.extend(res["words"])
    return words

transcript = transcribe_audio("/Data/aryanCodes3/noobies.ai/video/voice.mp3")
```
